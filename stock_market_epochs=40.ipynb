```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Tickers to analyze
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN']

# Settings
window_size = 60
forecast_horizon = 7  # Predict 7 days ahead
epochs = 40
batch_size = 32

def prepare_data(df):
    df = df[['Close', 'High', 'Low', 'Volume']]
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(df)

    X, y = [], []
    for i in range(window_size, len(scaled_data) - forecast_horizon + 1):
        X.append(scaled_data[i - window_size:i])
        y.append([scaled_data[i + j][0] for j in range(forecast_horizon)])

    X, y = np.array(X), np.array(y)

    split = int(0.8 * len(X))
    X_train, X_test = X[:split], X[split:]
    y_train, y_test = y[:split], y[split:]

    return X_train, X_test, y_train, y_test, scaler

def build_model(input_shape):
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=False, input_shape=input_shape))
    model.add(Dense(forecast_horizon))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

def inverse_transform(batch, scaler):
    padded = np.zeros((len(batch), 4))
    padded[:, 0] = batch
    return scaler.inverse_transform(padded)[:, 0]

# Initialize results list for saving
results = []

for ticker in tickers:
    print(f"\nüìà Processing {ticker}...\n")
    df = yf.download(ticker, start='2020-01-01', end='2024-12-31')
    X_train, X_test, y_train, y_test, scaler = prepare_data(df)

    model = build_model((X_train.shape[1], X_train.shape[2]))
    model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1)

    # Predict
    predicted = model.predict(X_test)

    # Calculate RMSE and MAPE for 1-day and 7-day predictions
    rmse_list, mape_list = [], []

    for day in range(forecast_horizon):
        true_vals = inverse_transform(y_test[:, day], scaler)
        pred_vals = inverse_transform(predicted[:, day], scaler)
        rmse = np.sqrt(mean_squared_error(true_vals, pred_vals))
        mape = mean_absolute_percentage_error(true_vals, pred_vals) * 100
        rmse_list.append(rmse)
        mape_list.append(mape)

        # Plot for each day
        plt.figure(figsize=(10, 4))
        plt.plot(true_vals, label='True Price')
        plt.plot(pred_vals, label='Predicted Price')
        plt.title(f'{ticker} - Day {day + 1} Ahead Forecast')
        plt.xlabel('Time')
        plt.ylabel('Price')
        plt.legend()
        plt.tight_layout()
        plt.show()

    # Store results for the current ticker
    avg_rmse_1day = rmse_list[0]  # RMSE for 1-day
    avg_mape_1day = mape_list[0]  # MAPE for 1-day
    avg_rmse_7day = np.mean(rmse_list)  # Average RMSE for 7-day
    avg_mape_7day = np.mean(mape_list)  # Average MAPE for 7-day

    results.append([ticker, avg_rmse_1day, avg_mape_1day, avg_rmse_7day, avg_mape_7day])

    print(f"üìä {ticker} RMSE (1-day): {avg_rmse_1day:.4f}")
    print(f"üìä {ticker} MAPE (1-day): {avg_mape_1day:.2f}%")
    print(f"üìä {ticker} RMSE (7-day): {avg_rmse_7day:.4f}")
    print(f"üìä {ticker} MAPE (7-day): {avg_mape_7day:.2f}%")

# Create a DataFrame to display the results
results_df = pd.DataFrame(results, columns=['Ticker', '1-Day RMSE', '1-Day MAPE (%)', '7-Day RMSE', '7-Day MAPE (%)'])

# Save the table as a CSV file
results_df.to_csv('stock_price_predictions_results.csv', index=False)

# Display the table in the notebook
print("\nüìä Evaluation Results (RMSE & MAPE):")
print(results_df)
```
